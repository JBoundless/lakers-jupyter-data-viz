```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import mpld3
df = pd.read_csv("/Users/jordanpagkalinawan/Desktop/NBA DV PROJECT/LAKER LEADERS/201920LakersRoster.csv")
playerName = df[df.Player == 'Player']
pointsPerGame = df[df.PTS == 'PTS']
reboundsPerGame = df[df.REB == 'REB']
assistsPerGame = df[df.AST == 'AST']
stealsPerGame = df[df.STL == 'STL']
blocksPerGame = df[df.BLK == 'BLK']
turnoversPerGame = df[df.TOV == 'TOV']
plt.title("2019-20 Lakers Points Per Game")
plt.bar(df.Player, df.PTS, color="gold")
plt.xticks(rotation = 'vertical')
plt.show();
fig = plt.figure();
html_str = mpld3.fig_to_html(fig)
Html_file= open("LALViz.html","w")
Html_file.write(html_str)
Html_file.close()
```
