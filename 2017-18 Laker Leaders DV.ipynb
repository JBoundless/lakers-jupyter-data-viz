```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv("/Users/jordanpagkalinawan/Desktop/NBA DV PROJECT/LAKER LEADERS/201718LakersRoster.csv")
playerName = df[df.Player == 'Player']
pointsPerGame = df[df.PTS == 'PTS']
reboundsPerGame = df[df.REB == 'REB']
assistsPerGame = df[df.AST == 'AST']
stealsPerGame = df[df.STL == 'STL']
blocksPerGame = df[df.BLK == 'BLK']
turnoversPerGame = df[df.TOV == 'TOV']
plt.title("2017-18 Lakers Points Per Game")
plt.bar(df.Player, df.PTS)
plt.xticks(rotation = 'vertical')
plt.show()
```
